# Data Preparation

#### MatchProjection(hazard,ear)
given both hazard and ear it will check both's projection system and convert if they differ


#### ClassifyHazard(hazard_file,base,stepsize)   
Classification of continue type of data to classified for loss assessment. Mostly intensity data can be classified with this function. base is starting point of classification and stepsize is interval. The classification will be done via formula:: base+i*stepsize where i is nth number of class

#### ProjectVector(ear,epsg)
Converts projection system of vector data to any given EPSG number

#### ProjectRaster(hazard,epsg)
Converts projection system of raster data to any given EPSG number

#### CheckProjectionRaster(hazard)
Handy tool to see the EPSG ID of raster data

#### CheckProjectionVector(ear)
Handy tool to see the EPSH ID or vector data

#### CheckUniqueTypes(EAR,Type Column)
Gives the unique types of values present in any column of a particular raster

#### LinkVulnerability(EAR,Type Column,vulnerabiliy_type)
linking code for type of EAR to vulnerability files.

#### ComputeCentroid(ear)
Computes the centroid of given polygon file.

# Exposure

Functions in EAR are listed as below

#### Exposure.ComputeExposure(ear,hazard,ear_key,outputfile,outputformat)
calculate exposure without aggregation in admin unit given the input of ear, hazard, the unique identification key in ear (Objectid) and outputfile and output format. output format can be either csv or shp

#### Exposure.ComputeExposureAgg(ear, hazard, ear_key, admin_unit, agg_col, outputname, outputformat='csv')

same as above but computes with aggregation in admin unit with admin unit and aggregation column in admin unit file 

Both of the functions can be run through batch process as stated in their tutorial

# Loss

#### Loss.computeLoss(exposure,ear,ear_key,costcol,vulncol,vulndir,haztype,outputfile,outputformat,               spprob=1,stepsize=1,base=0,hazunit=None,vulnunit=None,multiplyfactor=1,aggregation=False)

Loss computation function with input of exposure [either csv or shp] ear, ear unique id, cost column, vulnerability linked column, vulnerability file storage directory, hazard type [either Intensity or Susceptibility], output file and output format. further requirements optional are spatial probability, hazard classification step size, base, hazard unit [like meters], vulnerability unit ['like meters'], multiplyfactor to convert on the unit between hazard and vulnerability on the fly it will always convert hazard to vulnerability unit. such as if you have hazard in m and vulnerability in cm multiply factor should be 100. 

#### computeLossAgg(exposure,ear,ear_key,costcol,vulncol,vulndir,haztype,adminunit,admin_key,outputfile,outputformat,spprob=1,stepsize=1,base=0,hazunit=None,vulnunit=None,multiplyfactor=1,aggregation=True)

same as above but with aggregation true which will aggregate it to admin unit and also require two more parameters, admin unit and admin key which are admin unit file and its name for unique column.

# Risk Calculation
#### Risk.computeRisk(riskcombofile)
risk computation function for single hazard. the combinations file in csv format should be provided as shown in the template file. This do not aggregate risk to admin units

#### Risk.computeRiskAgg(riskcombofile)
Risk computation with aggregateion. input must be added in combination as in template. only for single hazard

#### Risk.ComputeRiskMh(riskcombofile,hazard_interaction)
It is for multi hazard risk analysis. requires two input files, riskcombination file as in template plus a hazard interaction file which specifies which hazard interacts with which and their probability in case of cascading and conditional hazard. the possible interactions are and method for computation are:



This function do not perform aggregation so you can leave the admin unit columns empty in template

#### Risk.ComputeRiskMhAgg(riskcombofile,hazard_interaction)

Same as above but with aggregation to admin units.
