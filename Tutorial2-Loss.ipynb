# RiskChanges Desktop : Loss

In loss Assessment, the major challenge is to manage the proper data. First and foremost, the EAR data have to have columns for Cost (or population or any value that you want to calculate risk) then the required vulnerability curve should be linked to each type of the EAR and EAR also must have the type of that particular element at risk. In the following Sections I will show how to prepare such columns and fit it into the template table.

## Linking Vulnerability curve to the Element at Risk

There must be a folder called Vulnerability in your workspace and each vulnerability file must be saved as *.csv file, the file name can be anything you want. once you put your vulnerability curve in those files be careful how do you store the vulnerability curves, the columns should look like this  where you can put Id as sequence of numbers and vuln_fk_id as null or just 1 because it comes form the server version withotut that it will not work. 

Now your Element at risk file must have Three major columns 
* **Type Column to define type of EAR such as RCC, Wood etc** * 
* ** Cost column to define cost of the EAR, it can be population cost of money or just number** *
* ** For different vulnerability curves create a new column such as Vuln_flood Vuln_eq etc and for each type of the EAR put the name of file that stores such vulnerability information** *

After you link it the EAR Table should look like this. The columns marked in red are important, but the name can be anything you want there is no restrictions in column name. Vulnerability curve name must be the same name as file. 



## Filling the Template

After you complete above step, now it is the time to fill the template, the existing template for this is quite big so you can keep consistent naming convention so that it will be easier to fill it using excel. The template for Loss analysis looks like this.



Here all last 5 columns are very similar to exposure. The EAR should have directory to your EAR file then, Exposure should link to a csv file of your exposure in relative or absolute directory. After that, Hazard Base and Hazard Stepsize are two values that you use to classify hazard in data standardization, it is the starting value and step size to classify in different calsses of your hazard. Hazard type represents either the data is intensity of susceptibility and CostColumn, Vulnerability Column and type Column are as specified above. 


Once you fill the Template we are ready for calculation and you can use the same way as exposure to perform the calculation.
